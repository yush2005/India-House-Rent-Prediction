Feature Engineering 


```python
df["bhk"] = df["house_type"].str.extract(r'(\d+)').astype(float)
df["bhk"] = df["bhk"].fillna(df["beds"])
```

```python
df=df.drop(['house_type'],axis=1)
```

```python
x=df.drop('rent',axis=1)
y=df['rent']                    
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
cat_cols=["city","locality","furnishing",]
num_cols=["area", "beds", "bathrooms", "balconies", "area_rate", "bhk"]
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder,StandardScaler

preprocessor=ColumnTransformer(
    transformers=[('num',StandardScaler(),num_cols),
                 ('cat',OneHotEncoder(handle_unknown='ignore'),cat_cols) ] 
)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor

model=RandomForestRegressor(n_estimators=200, random_state=42)

pipe=Pipeline(steps=[('preprocessor',preprocessor),
                     ('model',model)
            ]
)

pipe.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

y_pred = pipe.predict(x_test)
y_pred

print("MAE :", mean_absolute_error(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("R2  :", r2_score(y_test, y_pred))

```

```python
import joblib

joblib.dump(pipe, "house_rent_model.pkl")
print("Model saved successfully")
```

```python
df_reco = df.copy()
```

```python
def recommend_houses(
    df,
    model,
    city,
    budget,
    beds,
    min_area=0,
    furnishing=None,
    top_n=5
):
    filtered_df=df[df["city"]==city]

    filtered_df=filtered_df[filtered_df["beds"]==beds]

    if min_area > 0:
        filtered_df=filtered_df[filtered_df["area"] >= min_area]

    if furnishing:
        filtered_df=filtered_df[filtered_df["furnishing"]==furnishing]

    if filtered_df.empty:
        return pd.DataFrame()
    


    X_filtered = filtered_df.drop("rent", axis=1)
    filtered_df = filtered_df.copy()
    filtered_df["predicted_rent"] = model.predict(X_filtered)


    recommended = filtered_df[
        filtered_df["predicted_rent"] <= budget
    ]

    # 4ï¸âƒ£ Sort by best value (lowest predicted rent)
    recommended = recommended.sort_values(
        by="predicted_rent"
    )

    return recommended.head(top_n)

   
```

```python
results = recommend_houses(
    df=df_reco,
    model=pipe,
    city="Pune",
    budget=20000,
    beds=2,
    min_area=800,
    furnishing="Furnished",
    top_n=5
)


results[[
    "locality",
    "area",
    "beds",
    "furnishing",
    "predicted_rent"
]]
```
